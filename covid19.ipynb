```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://covid.ourworldindata.org/data/owid-covid-data.csv'
df = pd.read_csv(url)
df.head()
```

```python
countries = ['India', 'United States', 'Brazil']
df_filtered = df[df['location'].isin(countries)]
```

```python

plt.figure(figsize=(12, 6))

for country in countries:
    country_df = df_filtered[df_filtered['location'] == country]
    plt.plot(country_df['date'], country_df['total_cases'], label=country)
plt.title('COVID-19 Total Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

for country in countries:
    country_df = df_filtered[df_filtered['location'] == country]
    plt.plot(country_df['date'], country_df['new_cases'], label=country)

plt.title('yealy New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
